## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
```

## Dataframe

```python
df = pd.read_csv("06-wine.txt", sep="\t", index_col=False)
X = df.values
df
```

## Features

```python
col_id = 0
for col in df.columns:
    print(col_id,"\t", col)
    col_id+=1
```

## Normalazing

```python
X_mean = X.mean()
X_std = X.std()

X = (X - X_mean) / X_std
```

## Ploting

```python
plt.rcParams["figure.figsize"] = (16,8)
plt.axes(projection='3d').scatter3D(X[:,13], X[:,11], X[:,12], c=X[:,0])
plt
```

## Searching The Number Of Clusters

```python
inertia = []
disp = []

K = range(1, 12)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    clusters = kmeanModel.fit_predict(X)
    #inertia
    inertia.append(kmeanModel.inertia_)
    #sum of obj distances
    cluster_set = np.unique(clusters)
    sum_of_dist =  0
    
    for i in range(len(cluster_set)):
        cluster_mask = clusters[:] == cluster_set[i]
        sum_of_dist += pdist(X[cluster_mask,:]).sum()
    disp.append(sum_of_dist)
```

```python
plt.plot(K, disp, 'o-')
plt.xlabel('k')
plt.ylabel('Sum of obj distances')
plt
```

```python
plt.plot(K, inertia, 'o-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt
```

## Predicted Class Labels

```python
kmeanModel = KMeans(n_clusters=3)
clusters = kmeanModel.fit_predict(X)
clusters
```

```python
plt.axes(projection='3d').scatter3D(X[:,13], X[:,11], X[:,12], c=clusters)
plt
```
