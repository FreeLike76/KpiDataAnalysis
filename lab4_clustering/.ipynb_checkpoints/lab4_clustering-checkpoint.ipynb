## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform, cdist
from scipy.cluster.hierarchy import dendrogram, linkage, cophenet
from sklearn.cluster import AgglomerativeClustering
```

## Dataframe

```python
df = pd.read_csv("06-wine.txt", sep="\t", index_col=False)
X = df.values
df
```

## Features

```python
col_id = 0
for col in df.columns:
    print(col_id,"\t", col)
    col_id+=1
```

## Normalazing

```python
X_mean = X.mean()
X_std = X.std()

X = (X - X_mean) / X_std
```

## Ploting

```python
plt.rcParams["figure.figsize"] = (16,8)
plt.axes(projection='3d').scatter3D(X[:,13], X[:,11], X[:,12], c=X[:,0])
plt
```

## Distance. Linkage. Cophenet

```python
linkage_eval = {}

for linkage_method in ["single", "median", "centroid", "average"]:
    for linkage_metric in ["euclidean", "cityblock", "minkowski"]:
        if (linkage_method == "median" or linkage_method == "centroid") and linkage_metric != "euclidean":
            continue
        else:
            z = linkage(pdist(X, metric=linkage_metric, p=4), method=linkage_method)
            linkage_eval[linkage_method+"-"+linkage_metric] = max(cophenet(z))

linkage_eval = dict(sorted([(k, v) for k, v in linkage_eval.items()], key = lambda x: x[1], reverse=True))

for key in linkage_eval.keys():
    print(key, "\t", linkage_eval[key])

plt.bar(linkage_eval.keys(), linkage_eval.values())
```

```python
distance = pdist(X, metric="euclidean")
distance_matrix = squareform(distance)

plt.axes(projection='3d').contour3D(range(distance_matrix.shape[0]), range(distance_matrix.shape[1]), distance_matrix, 50)
```

```python
dend = dendrogram(linkage(distance, method="median"))
plt.show()
```

## Clustering

```python
cluster = AgglomerativeClustering(n_clusters=3, affinity="euclidean", linkage="average")
clusters = cluster.fit_predict(X)
plt.axes(projection='3d').scatter3D(X[:,13], X[:,11], X[:,12], c=clusters)
plt
```

## Predicted Class Labels

```python
clusters
```

## Centers and Dispersion

```python
center = []
disp = []

cluster_set = np.unique(clusters)
for i in range(len(cluster_set)):
    cluster_mask = clusters[:] == cluster_set[i]
    center.append([X[cluster_mask,j].mean() for j in range(X.shape[1])])
    disp.append(cdist([center[i]], X[cluster_mask,:]).mean())
```
